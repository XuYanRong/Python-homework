```python
# Import the pathlib and csv library 
from pathlib import Path
import csv

# Set the file path
csvpath = Path('../Resources/budget_data.csv')
budget_data = {}
# Create a output file
file_to_output = "budget_analysis.txt"

# Initialise all variables
total_months = 0 
total_pnl = 0  
previous_row = 0
pnl_change = 0
sum_of_change = 0
greatest_increase = 0
greatest_decrease = 0
pnl_change = 0
change= 0

# Open the csv file as a file object
with open(csvpath, "r") as csvfile:
    # Pass in the csv file to the csv.reader() function
    # (with ',' as the delmiter/separator) and return the csvreader object
    csvreader = csv.reader(csvfile, delimiter=',')
    
    # Go to the next row from the start of file 
    header = next(csvreader)
    

    
# Reader each row of data after the header
    for row in csvreader:
        month = row[0]
        pnl = int(row[1])
        budget_data[month] = int(pnl)  # Since the pnl on the list is still a string, then, convert from string to integer
        
    for  month, pnl, in budget_data.items(): # Recognise each row of the date and pnl are pairs
        total_months = total_months + 1 # The no. of months is counted by adding one for each row
        total_pnl = total_pnl + pnl # The total Profit/Loss is calculated by adding the next pnl once
    
        
  # To index the keys of a dictionary, must establish a new list that contains all the data       
    values = budget_data.values() 
    value_list = list(values)
    keys_list = list(budget_data)
    
   # Then, index i from the first row to the last row of the list to calculate the differences between two conservative periods
    for i in range(1,len(value_list)):
        sum_of_change += value_list[i]-value_list[i-1] # The cumulative differences is by the current pnl minus the previous pnl and add the previous changes 
        change = value_list[i]-value_list[i-1] # The difference of two consecutive periods    
        
    # The if condition helps to capture the greatest increase and decrease
        if change > greatest_increase: 
            greatest_increase = change
            Date1 = keys_list[i] # To show the corresponding date of the greatest increase
        elif change < greatest_decrease: 
            greatest_decrease = change 
            Date2 = keys_list[i] 
        
            
# Store the information of current row to previous_row 
previous_row = row
average_change = sum_of_change / (total_months - 1)           
            
            
print("Financial Analysis")            
print("----------------------------")            
print("Total Month: ", total_months)
print("Total: $", total_pnl)
print("Average of change: $", average_change)
print("Greatest Increase in Profits:", Date1 , greatest_increase)
print("Greatest Decrease in Profits:", Date2 , greatest_decrease)
```
